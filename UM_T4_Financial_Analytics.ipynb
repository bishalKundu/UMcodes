```
#Importing Libraries

import pandas as pd
import seaborn as sns

#Importing Data

d1=pd.read_csv('/content/Financial Analytics data.csv')
d1.head()
```

```
#Checking Null Values

d1.isna().sum()
```

```
#Removing the Null Values

d1.dropna(inplace=True)
d1.isna().sum()
```

```
#Checking for Duplicate Values

d1['Name'].duplicated().sum()
```

```
#Checking Data Info and Descriptive Statistics of the data

d1.info()
d1.describe()
```

```
#Removing 'S.No.' Column as it's not needed

del d1['S.No.']

#Adding new columns for Annual sales and Price-to-Sales Ratio

d1['Estimated Sales/yr - Crore']=4*d1['Sales Qtr - Crore']
d1['P/S Ratio']=d1['Mar Cap - Crore']/d1['Estimated Sales/yr - Crore']
d1.head()
```

```
#Checking for outliers

d1.boxplot(figsize=(15,7))
d1.describe()
```

```
#Histogram before Removing the outliers

sns.histplot(d1)
```

```
#Finding IQR (Inter Quartile Range), UW(Upper Whisker)

MC_IQR=23549.900-4843.575     #Mar Cap
MC_UW=23549.900+1.5*(MC_IQR)

SQ_IQR=2730.195-534.910       #Sales/Qtr
SQ_UW=2730.195+1.5*(SQ_IQR)

ES_IQR=10920.780-2139.640     #Sales/Yr
ES_UW=10920.780+1.5*(ES_IQR)

PS_IQR=4.171-1.571            #P/S Ratio
PS_UW=4.171+1.5*(PS_IQR)


MC_UW,SQ_UW,ES_UW,PS_UW
```

```
#Adjuating the outliers with upper whisker values

d1.loc[d1['Mar Cap - Crore']>51609.387500000004,'Mar Cap - Crore']=51609.387500000004
d1.loc[d1['Sales Qtr - Crore']>6023.122500000001,'Sales Qtr - Crore']=6023.122500000001
d1.loc[d1['Estimated Sales/yr - Crore']>24092.490000000005,'Estimated Sales/yr - Crore']=24092.490000000005
d1.loc[d1['P/S Ratio']>8.071000000000002,'P/S Ratio']=8.071000000000002

#Updated boxplot

d1.boxplot(figsize=(15,7))
```

```
#Updated Histogram with Adjusted Outliers

sns.histplot((d1['Mar Cap - Crore'],d1['Sales Qtr - Crore'],d1['P/S Ratio']))
```

```
#Corelation of the data

Correl=d1['Mar Cap - Crore'].corr(d1['Estimated Sales/yr - Crore'])
print('Corelation=',Correl)
print('R Square=',Correl**2)
```
